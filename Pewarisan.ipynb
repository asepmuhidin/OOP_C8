```python
class Pegawai:
  def __init__(self,nama,gaji):
    self.nama=nama
    self.gaji=gaji
    print("Ini menjalankan method konstrktor kelas pegawai")

  def info(self):
    print(f"Nama {self.nama}, gaji : {self.gaji}")

  def jamLembur(self,jamAktual):
    return 1.5*jamAktual

  def upahLembur(self,jamAktualLembur):
    return (1/173)*self.gaji*self.jamLembur(jamAktualLembur)
```

```python
#Membuat objek p1 dari kelas pegawai
p1=Pegawai("Bagus",5800000)
p1.info()
aktualLembur=3
upahLembur=p1.upahLembur(aktualLembur)
print(f"Upah {aktualLembur} jam lembur {p1.nama} dengan gaji pokok {p1.gaji:,} adalah {upahLembur:,.2f}")
```

```python
class Supervisor(Pegawai):
  def __init__(self,nama,gaji,tunjangan):
    super().__init__(nama,gaji)
    self.tunjangan=tunjangan
    print("Ini menjalankan method konstrktor kelas Supervisor")

  def upahLembur(self, jamAktualLembur):
    if super().jamLembur(jamAktualLembur)>=3:
      return super().upahLembur(jamAktualLembur)
    else:
      return 0
```

```python
#Membuat objek supervisor spv1 dari kelas supervisor

spv1=Supervisor("Budi",6500000,500)
aktualLembur=2
upahLembur=spv1.upahLembur(aktualLembur)
print(f"Upah {aktualLembur} jam lembur {spv1.nama} dengan gaji pokok {spv1.gaji:,} adalah {upahLembur:,.2f}")
```

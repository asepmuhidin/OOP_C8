## Method Overriding

```python
class Hewan:
  def __init__(self,nama):
    self.nama=nama
  def suara(self):
    return "Suara hewan"
```

```python
class Anjing(Hewan):
  def suara(self):
    return "guk..guk"

class Kucing(Hewan):
  def suara(self):
    return "Meoongg.."

class Sapi(Hewan):
  def suara(self):
    return "Moooohh"
```

```python
anjing=Anjing("Buddy")
kucing=Kucing("Wheasker")
sapi=Sapi("Bessie")
print(f"{anjing.nama} bersuara {anjing.suara()}")
print(f"{kucing.nama} bersuara {kucing.suara()}")
print(f"{sapi.nama} bersuara {sapi.suara()}")
```

```python
animals=[
    Anjing("Sloki"),
    Kucing("Ucup"),
    Sapi("Jamal")
]
for animal in animals:
  print(f"{animal.nama} bersuara {animal.suara()}")
```

## Duck typing

```python
class Mobil:
  def start(self):
    return "mesin mobil menyala"

class Motor:
  def start(self):
    return "mesin motor menyala"

class Perahu:
  def start(self):
    return "mesin perahu menyala"

def operasikan_kendaraan(kendaraan):
  return print(kendaraan.start())
```

```python
daftar_kendaraan=[Mobil(),Motor(),Perahu(),Motor()]
for kendaraan in daftar_kendaraan:
  operasikan_kendaraan(kendaraan)
```

## Operator Overloading

```python
class Apple:
  def __init__(self,jumlah):
    self.jumlah=jumlah

  def __add__(self,other):
    return Apple(self.jumlah+other.jumlah)

  def __str__(self):
    return f"Jumlah apel {self.jumlah}"
```

```python
apel1=Apple(12)
apel2=Apple(11)
apel3=apel1+apel2

print(apel3)
```

# Akses Modifier (Public, Private, Protected)

## Public

```python
class Mahasiswa:
  def __init__(self,nama):
    self.nama=nama

  def sayHello(self):
    print(f"Hello {self.nama}")
```

```python
mhs1=Mahasiswa("Abdan Rizky")
mhs1.sayHello() # ini method public
print(f"Haloo juga {mhs1.nama}") # ini properti public
```

## Protected

```python
class Mahasiswa:
  def __init__(self,nama):
    self._nama=nama

  def _sayHello(self):
    print(f"Hello {self._nama}")
```

```python
#TIDAK dianjurkan mengakses prop/method dari luar kelas
#walaupun di python masih bisa diakses
mhs1=Mahasiswa("Abdan Rizky")
mhs1._sayHello() # ini method protected
print(f"Haloo juga {mhs1._nama}") # ini properti protected
```

## Private

```python
class Mahasiswa:
  def __init__(self,nama):
    self.__nama=nama

  def __sayHello(self):
    print(f"Hello {self.__nama}")
```

```python
mhs1=Mahasiswa("Candra")
#print(f"Haloo .. {mhs1.__nama}") -->Tidak bisa diakses langsung

#Menggunakan name mangling dengan aturan (_NamaKelas__atribute/method)
print(f"Haloo {mhs1._Mahasiswa__nama}")
mhs1._Mahasiswa__sayHello()
```

# Getter & Seetter

```python
class Mahasiswa:
  def __init__(self,nim,nama):
    self.__nim=nim
    self.__nama=nama

  #Getter nim
  def get_nim(self):
    return self.__nim
  #Setter nim
  def set_nim(self,nim):
    self.__nim=nim

  #Getter nama
  def get_nama(self):
    return self.__nama
  #Setter nama
  def set_nama(self,nama):
    self.__nama=nama
```

```python
mhs1=Mahasiswa("1234","Aldi Wahyu")
```

```python
#print(f"Haloo {mhs1.__nama}")-->ERROR karena private attribut nama
print(f"Haloo {mhs1.get_nama()}")
```

```python
#mhs1.__nama="Dimas Andika"
mhs1.set_nama("Dimas Andika")
print(f"Haloo {mhs1.get_nama()}")
```

## Getter - Setter Python

```python
class Mahasiswa:
  def __init__(self,nim,nama):
    self.__nim=nim
    self.__nama=nama

  @property
  def nim(self):
    return self.__nim

  @property
  def nama(self):
    return self.__nama

  @nama.setter
  def nama(self,nama_baru):
    self.__nama=nama_baru
```

```python
mhs1=Mahasiswa("123","Jamaludin")
print(f"Hallo {mhs1.nama}")
mhs1.nama="Gulam Fikri"
print(f"Hallo {mhs1.nama}")
```

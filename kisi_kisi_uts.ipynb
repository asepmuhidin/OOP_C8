```python
# Kelas induk
class Hewan:
    def __init__(self, nama, jenis, berat):
        self.nama = nama
        self.jenis = jenis
        self.berat = berat

    def hitung_makan(self):
        # Akan di-override oleh subclass
        pass

    def tampilkan_info(self):
        print(f"Hewan: {self.nama} ({self.jenis})")
        print(f"Berat: {self.berat} kg")
```

```python
# Kelas turunan: Karnivora
class Karnivora(Hewan):
    def __init__(self, nama, berat, jumlah_daging):
        super().__init__(nama, "Karnivora", berat)
        self.jumlah_daging = jumlah_daging

    def hitung_makan(self):
        biaya = self.jumlah_daging * 80000
        return biaya


# Kelas turunan: Herbivora
class Herbivora(Hewan):
    def __init__(self, nama, berat, jumlah_rumput):
        super().__init__(nama, "Herbivora", berat)
        self.jumlah_rumput = jumlah_rumput

    def hitung_makan(self):
        biaya = self.jumlah_rumput * 15000
        return biaya


# Kelas turunan: Omnivora
class Omnivora(Hewan):
    def __init__(self, nama, berat, jumlah_makanan):
        super().__init__(nama, "Omnivora", berat)
        self.jumlah_makanan = jumlah_makanan

    def hitung_makan(self):
        biaya = self.jumlah_makanan * 50000
        return biaya
```

```python
# ============================================================
# Contoh penggunaan program
# ============================================================

# Membuat objek dari masing-masing kelas
singa = Karnivora("Singa", 190, 5)
kuda = Herbivora("Kuda", 250, 10)
beruang = Omnivora("Beruang", 300, 5)

# Menampilkan informasi dan hasil perhitungan
daftar_hewan = [singa, kuda, beruang]

for hewan in daftar_hewan:
    hewan.tampilkan_info()
    print(f"Biaya makan per hari: Rp{hewan.hitung_makan():,.0f}")
    print("-" * 40)
```
